```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load your dataset
data = pd.read_csv('all_data.csv')

# Print the column names to check for the correct label column name
print(data.columns)

# Preprocess your data (this is an example, adjust as needed)
# Assume columns: 'B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI', 'Label'
features = data[['B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI']]
# Update the column name if needed after checking the printed columns
labels = data['label']  # Or data['LABEL'] or data[' Label'], etc.

# Handle missing values (NaN) in labels:
# You can either drop rows with NaN labels or impute them with a suitable value.
# Here, we'll drop rows with NaN labels:
data = data.dropna(subset=['label'])  # Drop rows where 'label' is NaN

# Reset indices after dropping rows
data = data.reset_index(drop=True)

# Update features and labels after handling NaN values
features = data[['B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI']]
labels = data['label']

# Convert labels to integers and ensure they start from 0:
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)  # Fit and transform labels

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# Initialize XGBoost classifier
model = XGBClassifier(learning_rate=0.1, max_depth=5, n_estimators=100)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'Accuracy: {accuracy * 100:.2f}%')
print('Confusion Matrix:')
print(conf_matrix)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import torch  # Import torch to check for cuda availability

# Load your dataset
data = pd.read_csv('all_data.csv')

# Print the column names
print(data.columns)

# Preprocess your data
features = data[['B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI']]
labels = data['label']

# Handle missing values in labels
data = data.dropna(subset=['label'])
data = data.reset_index(drop=True)

# Update features and labels
features = data[['B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI']]
labels = data['label']

# Convert labels to integers
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# Check for GPU availability using torch
device = 'GPU' if torch.cuda.is_available() else 'CPU'
# or simply device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Define models and hyperparameter grids
models = {
    'CatBoost': (CatBoostClassifier(verbose=0, early_stopping_rounds=10,
                                   task_type='GPU' if 'GPU' in str(device) else 'CPU'), {
        # Use GPU if available
        'iterations': [100, 200],
        'depth': [4, 6],
        'learning_rate': [0.01, 0.1]
    }),
    'LightGBM': (LGBMClassifier(early_stopping_rounds=10,
                                 device_type='gpu' if 'GPU' in str(device) else 'cpu'), {
        # Use GPU if available
        'n_estimators': [100, 200],
        'max_depth': [4, 6],
        'learning_rate': [0.01, 0.1]
    }),
    'RandomForest': (RandomForestClassifier(), {
        'n_estimators': [100, 200],
        'max_depth': [4, 6]
    }),
    'GradientBoosting': (GradientBoostingClassifier(), {
        'n_estimators': [100, 200],
        'max_depth': [4, 6]
    })
}

# Loop through the models and evaluate
best_accuracy = 0
best_model = None  # Initialize best_model

for model_name, (model, param_grid) in models.items():
    print(f"Training {model_name}...")

    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy',
                               n_jobs=-1)  # Reduced cv to 3

    # Pass eval_set to LightGBM for early stopping
    if model_name == 'LightGBM':
        grid_search.fit(X_train, y_train, eval_set=[(X_test, y_test)], eval_metric='logloss')
    else:
        grid_search.fit(X_train, y_train)

    best_model_instance = grid_search.best_estimator_

    y_pred = best_model_instance.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'{model_name} Accuracy: {accuracy * 100:.2f}%')
print(f'{model_name} Confusion Matrix:')
print(conf_matrix)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
import torch  # Import torch to check for cuda availability

# Load your dataset
data = pd.read_csv('all_data.csv')

# Print the column names
print(data.columns)

# Preprocess your data
features = data[['B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI']]
labels = data['label']

# Handle missing values in labels
data = data.dropna(subset=['label'])
data = data.reset_index(drop=True)

# Update features and labels
features = data[['B3', 'B5', 'B8', 'NDWI', 'NDVI', 'FDI']]
labels = data['label']

# Convert labels to integers
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# Check for GPU availability using torch (not used by RandomForest but kept for consistency)
device = 'GPU' if torch.cuda.is_available() else 'CPU'

# Define the RandomForest model and hyperparameter grid
model = RandomForestClassifier()
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6]
}

# Perform GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_  # Get the best model

# Make predictions using the best model
y_pred = best_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f'RandomForest Accuracy: {accuracy * 100:.2f}%')
print('RandomForest Confusion Matrix:')
print(conf_matrix)
```
